```python
df['dteday'] = pd.to_datetime(df['dteday'])
df['dteday'] += pd.to_timedelta(df['hr'], unit='h')
df.index = df['dteday']
del df['dteday']
```

```python
df.head()
```

```python
df.columns = ['Instance','Season','Year','Month','Hour','Holiday','Weekday','Workingday','Weather_Condition','Normalized_Temperature','Normalized_Feels_Temperture','Huumidity','Windspeed','Casual','Registered','Demand']
```

```python
df.head()
```

```python
df["Workingday"] = np.where(df["Workingday"] == 1, True, False)
```

```python
df.head()
```

```python
df.groupby('Weekday').Demand.sum().plot()
```

```python
cd = df[df['Casual']>df['Registered']].groupby('Weekday').Demand.sum()
cd
```

```python
cd.plot()
```

```python
df['Part_Of_Day'] = pd.cut(np.array(df['Hour']),3, labels=[0, 1, 2])
df.head()
```

```python
df.groupby('Part_Of_Day').Demand.sum().plot()
```

```python
pf = df.pivot(columns='Weather_Condition',values='Demand')
pf.head()
```

```python
pf = pf.fillna(0)
pf.head()
```
