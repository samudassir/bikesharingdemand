```python
df.columns = ['Instant','Date','Season','Year','Month','Hour','Holiday','Weekday','Workingday','Weather_Condition','Normalized_Temperature','Normalized_Feels_Temperture','Huumidity','Windspeed','Casual','Registered','Demand']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] += pd.to_timedelta(df['Hour'], unit='h')
df.index = df['Date']
del df['Date']
del df['Instant']
```

```python
df.info()
```

```python
#df["Workingday"] = np.where(df["Workingday"] == 1, True, False)
```

```python
cd = df[df['Casual']>df['Registered']]
plt.bar(cd['Weekday'],cd['Casual'],color='c')
plt.bar(cd['Weekday'],cd['Registered'],color='y')
plt.xlabel('Weekday')
plt.ylabel('Demand For Bikes')
plt.title('Days with higher Casual Demand')
plt.legend(labels=["Casual", "Registered"])
savefig('../../img/plot/casual_demand_greaterthan_registered_demand.png', bbox_inches='tight')
plt.show()
```

```python
df['Part_Of_Day'] = pd.cut(np.array(df['Hour']),6, labels=[0,1,2,3,4,5])
df.head()
```

```python
sns.set_style("white")
r = np.arange(7)
reg = df.groupby(['Weekday']).Registered.mean()
cas = df.groupby(['Weekday']).Casual.mean()
barWidth = 0.85
opacity = 0.4
totals = [i+j for i,j in zip(reg, cas)]
registered = [i / j * 100 for i,j in zip(reg, totals)]
casual = [i / j * 100 for i,j in zip(cas, totals)]

plt.bar(r, registered, bottom=casual, color='orange', alpha=opacity,edgecolor='white', width=barWidth)
plt.bar(r, casual, color='green', edgecolor='white', alpha=opacity,width=barWidth)

plt.xticks(r,('Sun','Mon','Tue','Wed','Thu','Fri','Sat'))
plt.xlabel('Weekday')
plt.ylabel('Demand For Bikes in %')
plt.title('Casual and Registered Weekday Demand in %')
plt.legend(labels=["Registered","Casual"],loc=2,bbox_to_anchor=(1,1),ncol=1)
savefig('../../img/plot/mean_casual_registered_demand_percent.png', bbox_inches='tight')
plt.show()
```

```python
fig, ax = subplots()

index = np.arange(7)
bar_width = 0.35
opacity = 0.4
ax.bar(index, df.groupby(['Weekday']).Registered.mean(), bar_width,
                alpha=opacity, color='b')
ax.bar(index + bar_width, df.groupby(['Weekday']).Casual.mean(), bar_width,
                alpha=opacity, color='r')

ax.set_xlabel('Weekday')
ax.set_ylabel('Demand For Bikes')
ax.set_title('Casual and Registered Weekday Demand')
ax.set_xticklabels(('Sun','Mon','Tue','Wed','Thu','Fri','Sat'))
ax.legend(labels=["Registered","Casual"],loc=2)
ax.set_xticks(index + bar_width / 2)
plt.xticks(rotation=45)
fig.tight_layout()
savefig('../../img/plot/mean_casual_registered_demand.png', bbox_inches='tight')
plt.show()
```

```python
fig, ax = subplots()

index = np.arange(24)
bar_width = 0.35
opacity = 0.4
ax.bar(index, df.groupby(['Hour']).Registered.mean(), bar_width,
                 color='orange')
ax.bar(index + bar_width, df.groupby(['Hour']).Casual.mean(), bar_width,
                 color='green')

ax.set_xlabel('Hour')
ax.set_ylabel('Demand For Bikes')
ax.set_title('Casual and Registered Hourly Demand')
ax.legend(labels=["Registered","Casual"],loc=0)
ax.set_xticks(np.arange(24))
plt.xticks(rotation=45)
fig.tight_layout()
savefig('../../img/plot/mean_casual_registered_hourly_demand.png', bbox_inches='tight')
plt.show()
```

```python
sns.set_style("white")
fig, ax = subplots()
index = np.arange(6)
bar_width = 0.50
ax.bar(index, df.groupby(['Part_Of_Day']).Demand.mean(), bar_width,color='grey')
ax.set_xlabel('Period')
ax.set_ylabel('Average Demand For Bikes')
ax.set_title('Demand for Bikes during different hourly windows')
ax.set_xticklabels(('0-4','4-8','8-12','12-16','16-20','20-23'))
ax.set_xticks(index + bar_width/10)
fig.tight_layout()
savefig('../../img/plot/demand_in_parts_of_day.png', bbox_inches='tight')
plt.show()
```
